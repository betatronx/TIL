## 복소수
* Complex Numbers
* 출처 : [복소수 - 위키백과, 우리 모두의 백과사전](https://ko.wikipedia.org/wiki/%EB%B3%B5%EC%86%8C%EC%88%98)

수학에서의 복소수(複素數, complex number)는 다음 꼴로 나타낼 수 있는 수 이다.

{\displaystyle \,a+bi} \,a+bi
이 때 {\displaystyle a} a, {\displaystyle b} b는 실수이고 {\displaystyle i} i는 허수단위로 {\displaystyle i^{2}=-1} i^{2}=-1을 만족한다. 실수 {\displaystyle a} a를 그 복소수의 실수부, 실수 {\displaystyle b} b를 복소수의 허수부라고 부른다. 모든 실수는 복소수에 포함된다. 왜냐하면 모든 실수는 허수부가 0인 복소수로 표시할 수 있기 때문이다. 즉 실수 {\displaystyle a} a 는 복소수 {\displaystyle a+0i} a+0i 와 같다. 예를 들어, {\displaystyle {\sqrt {13}}} {\displaystyle {\sqrt {13}}}은 실수부가 {\displaystyle {\sqrt {13}}} {\displaystyle {\sqrt {13}}}이고 허수부가 0인 복소수이다.

복소수에서도 실수에서 성립하는 사칙 연산을 정의할 수 있고, 기존의 성질을 대부분 만족한다. 예를 들어, 복소수에서도 실수에서도 마찬가지로 사칙연산에 대해 닫혀 있다.

전자공학 등의 분야에서는 전류의 기호로 {\displaystyle i} i를 사용하기 때문에, 혼동을 피하기 위해 허수단위를 {\displaystyle j} j로 표기하는 경우도 있다

### 이산 푸리에 변환 Discrete Fourier Transform
## Q5.
a 의 1차원 DFT를 계산한다.<br/>
Compuete the one-dimensional DFT of a.<br/>

[2) 주파수 샘플링과 DFT - 공돌이의 수학정리노트](https://wikidocs.net/4046)

```python
a = np.exp(2j * np.pi * np.arange(8))
output = np.fft.fft(a)
output
```

## Q6.
위 문제 output의 1차원 역 DFT를 계산한다.<br/>
Compute the one-dimensional inverse DFT of the output in the above question.

```python
print("a=", a)
inversed = np.fft.ifft(output)
print("inversed=", a)
```

* 이산 푸리에 변환(discrete Fourier transform, DFT)은 이산적인 입력 신호에 대한 푸리에 변환으로, 디지털 신호 분석과 같은 분야에 사용된다.


```python
a = [0, 1, 0, 0]
output = np.fft.rfft(a)
print(output)
assert output.size==len(a) //2+1 if len(a)%2==0 else (len(a)+1) //2

# cf.
output2 = np.fft.fft(a)
print(output2)
```

## Q9.
a의 DFT 샘플빈도를 구한다.<br/>
Return the DFT sample frequencies of a.

```python
signal = np.array([-2, 8, 6, 4, 1, 0, 3, 5], dtype=np.float32)
fourier = np.fft.fft(signal)
n = signal.size
freq = np.fft.fftfreq(n, d=1)
freq
```

## Window Functions
[Window function - Wikipedia](https://en.wikipedia.org/wiki/Window_function)

```python
fig = plt.figure(figsize=(19,10))

# Hamming window
window = np.hamming(51)
plt.plot(np.bartlett(51), label="Bartlett window")
plt.plot(np.blackman(51), label="Blackman window")
plt.plot(np.hamming(51), label="Hamming window")
plt.plot(np.hanning(51), label="Hanning window")
plt.plot(np.kaiser(51, 14), label="Kaiser window")
plt.xlabel("sample")
plt.ylabel("amplitude")
plt.legend()
plt.grid()

plt.show()
```
